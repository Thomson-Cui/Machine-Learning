# 1. Data Collection

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Housing.csv")
m = len(data)
print("The number of data in the set:",m)
```

```python
data.head(10)
```

```python
price = data['Price'] / 1000
square = data['Square'] / 100

plt.scatter(square, price, s=1)
plt.title("Housing")
plt.xlabel("Square x 10^2")
plt.ylabel("Price x 10^4")
plt.show()
```

# 2. 

```python
def Predicting_Function(w, b, x):
    
    y = w * x + b
    
    return y

def Gradient_Computing(square, price, w, b):
    
    m = len(square)
    
    Gradient_w = 0
    
    Gradient_b = 0
    
    for i in range(m):
        
        Gradient_w = Gradient_w + (Predicting_Function(w, b, square[i]) - price[i])*square[i]
        
        Gradient_b = Gradient_b + (Predicting_Function(w, b, square[i]) - price[i])
        
    Gradient_w = Gradient_w / m
    
    Gradient_b = Gradient_b / m
    
    return Gradient_w, Gradient_b

def Cost(w, b, x, y):
    
    m = len(x)
    
    J_w_b = 0
    
    for i in range(m):
        
        J_w_b = J_w_b + (Predicting_Function(w, b, x[i]) - y[i])**2
    
    J_w_b = J_w_b / (2*m)
    
    return J_w_b

def Gradient_Descent(alpha, w, b, square, price, iterations):
    
    m = len(square)
            
    J_w_b = Cost(w, b, square, price)
    
    print("Iteration:0,w =",w,"b =",b,"J_w_b =",J_w_b)
    
    for i in range(iterations - 1):
    
        Gradient_w, Gradient_b = Gradient_Computing(square, price, w, b)

        temp_w = w - alpha * Gradient_w

        temp_b = b - alpha * Gradient_b

        w = temp_w

        b = temp_b
        
        J_w_b = Cost(w, b, square, price)
        
        print("Iteration: ",i,"w =",w,"b =",b,"J_w_b =",J_w_b)

    return w, b, J_w_b
```

```python
alpha = 0.01
w = 0 
b = 0
iterations = 1000

w, b, J_w_b = Gradient_Descent(alpha, w, b, square, price, iterations)

print("Outcome:",w ,b ,J_w_b)
```

```python
# Placeholder data
square = [1, 2, 3, 4, 5]
price = [2, 4, 5, 4, 5]

def Predicting_Function(w, b, x):
    y = w * x + b
    return y

def Gradient_Computing(square, price, w, b):
    m = len(square)
    Gradient_w = 0
    Gradient_b = 0
    for i in range(m):
        Gradient_w += (Predicting_Function(w, b, square[i]) - price[i]) * square[i]
        Gradient_b += (Predicting_Function(w, b, square[i]) - price[i])
    Gradient_w /= m
    Gradient_b /= m
    return Gradient_w, Gradient_b

def Cost(w, b, x, y):
    m = len(x)
    J_w_b = 0
    for i in range(m):
        J_w_b += (Predicting_Function(w, b, x[i]) - y[i])**2
    J_w_b /= (2 * m)
    return J_w_b

def Gradient_Descent(alpha, w, b, square, price, iterations):
    m = len(square)
    J_w_b = Cost(w, b, square, price)
    print("Iteration: 0, w =", w, "b =", b, "J_w_b =", J_w_b)
    for i in range(iterations - 1):
        Gradient_w, Gradient_b = Gradient_Computing(square, price, w, b)
        temp_w = w - alpha * Gradient_w
        temp_b = b - alpha * Gradient_b
        w = temp_w
        b = temp_b
        J_w_b = Cost(w, b, square, price)
        if i % 100 == 0:  # Print every 100 iterations
            print("Iteration: ", i, "w =", w, "b =", b, "J_w_b =", J_w_b)
    return w, b, J_w_b

alpha = 0.01
w = 0 
b = 0
iterations = 10000

w, b, J_w_b = Gradient_Descent(alpha, w, b, square, price, iterations)

print("Outcome:", w, b, J_w_b)
```
