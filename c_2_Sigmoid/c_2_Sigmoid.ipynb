## Sigmoid function

$$f_{\vec{\mathrm{w}},b}(\vec{\mathrm{x}})=\frac{1}{1+e^{-(\vec{\mathrm{w}}\cdot\vec{\mathrm{x}}+b)}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from plt_one_addpt_onclick import plt_one_addpt_onclick

def sigmoid(z):
    
    f = 1 / (1 + np.exp(-z))
    
    return f
```

```python
z_tmp = np.arange(-10, 11)

y = sigmoid(z_tmp)

np.set_printoptions(precision=3) 

print("Input (z), Output (sigmoid(z))")
print(np.c_[z_tmp, y])
```

```python
zeros_line_x = np.zeros(11)

zeros_line_y = np.arange(11) / 10

fig, ax = plt.subplots(1, 1, figsize = (5,3))

ax.plot(z_tmp, y, c = "blue")

ax.scatter(zeros_line_x, zeros_line_y, c = "black", s = 20, marker = "|")

ax.set_title("Sigmoid function")

ax.set_ylabel("sigmoid(z)")

ax.set_xlabel("z")

plt.show()
```
