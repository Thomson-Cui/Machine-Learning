# 高偏差（High Bias）

* 表现特征： 模型对训练数据和新数据的拟合能力较差。
* 原因： 通常是因为模型过于简单，不能很好地捕捉数据的复杂性。这可能是因为模型假设过于简化，或者特征选择不足。
* 结果： 在训练集和测试集上都表现不佳，模型未能充分学习数据的模式。

# 高方差（High Variance）

* 表现特征： 模型在训练数据上表现良好，但在新数据上表现较差。
* 原因： 通常是因为模型过于复杂，过度拟合了训练数据中的噪声和细节。这可能是因为模型具有过多的参数，或者训练数据不足。
* 结果： 在训练集上表现很好，但在测试集上可能过度拟合，泛化能力差。

## 关键区别：

偏差（Bias） 衡量了模型在训练数据上的拟合能力，高偏差表示模型无法很好地适应数据。
方差（Variance） 衡量了模型对训练数据的小波动的敏感程度，高方差表示模型对数据中的噪声过于敏感。
在实际应用中，找到合适的模型复杂度（偏差-方差权衡）是很重要的。通常，过于简单的模型可能导致高偏差，而过于复杂的模型可能导致高方差。机器学习算法的目标是在这两者之间找到一个平衡，以获得更好的泛化性能。



# 泛化（Generalization）

在机器学习领域，"generalization"（泛化）是指模型在面对未见过的数据时的表现能力。一个在训练数据上表现良好并且能够很好地适应新数据的模型被认为具有良好的泛化能力。

泛化的目标是使模型能够捕捉数据的普遍模式，而不仅仅是在训练数据上表现良好。一个具有良好泛化能力的模型能够从训练数据中学到通用规律，而不是过度依赖于训练数据中的特定细节或噪声。

在训练过程中，模型通过学习训练数据中的模式和关系，然后希望将这些学到的模式应用到新的、未见过的数据上。泛化的能力是一个模型成功应对新情况的关键因素，因为在实际应用中，模型通常会面对不同于训练数据的输入。

评估模型的泛化能力通常通过在训练集之外的测试集上进行性能评估来完成。一个模型如果在测试集上表现好，表明它能够泛化到新数据，而不仅仅是记住了训练数据。

总体而言，泛化是机器学习中一个核心的概念，关注的是模型在未见过的数据上的适应能力和预测能力。
